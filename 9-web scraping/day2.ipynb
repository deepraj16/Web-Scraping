```python
from bs4 import BeautifulSoup as bs
import requests
Link = 'https://www.imdb.com/title/tt0068646/review'
```

```python

page = requests.get(Link)
page #if 200 then request is grant
```

```python
page.content #https://m.imdb.com/title/tt0068646/reviews?ref_=tt_urv
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
title = soup.find_all('a',class_='title')
title
```

```python
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
len(review_title)
```

```python
review_title[:]=[title.strip('\n') for title in review_title]
```

```python
review_title
```

```python
len(review_title)
```

```python
### scrping the rateing from given form
rating = soup.find_all('span',class_='point-scale')
rating
```

```python
rate=[]
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
```

```python
rate
```

```python
rate[:] = [r.strip('/') for r in rate]
rate
```
